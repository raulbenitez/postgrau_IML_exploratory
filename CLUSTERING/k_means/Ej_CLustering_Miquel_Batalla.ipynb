```
import pandas as pd
import seaborn as sns
mpg=sns.load_dataset('mpg')
mpg=mpg.drop(columns=['cylinders','model_year'])
mpg_num=mpg.select_dtypes(include='number').dropna()
mpg_num.head()
```

```
sns.pairplot(mpg,hue="origin")
```

```
from sklearn.cluster import KMeans

#Cogemos 3 grupos distintos 
group = KMeans(n_clusters=3).fit_predict(mpg_num)
mpg_grouped = mpg_num
mpg_grouped['kmeans'] = group
g = sns.pairplot(mpg_grouped, hue="kmeans")
```

```
gr = KMeans(n_clusters=3).fit(mpg_num)
print(gr.labels_)
print(gr.cluster_centers_)
print(gr.n_iter_)
```
