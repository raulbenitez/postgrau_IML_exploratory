```
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.datasets import make_blobs
from sklearn.cluster import AgglomerativeClustering
from sklearn import mixture

random_state = 170

# Generate data:
n_samples = 1500
centers1 = [(0, 0), (7, 7)]
X1, y1 = make_blobs(n_samples=n_samples, centers=centers1, n_features=2,random_state=0)
transformation1 = [[0.60834549, -0.63667341], [-0.40887718, 0.85253229]]
X_aniso1 = np.dot(X1, transformation)

centers2 = [(0, 0)]
X2, y2 = make_blobs(n_samples=n_samples, centers=centers2, n_features=1,random_state=0)
transformation2 = [ [-0.40887718, -0.85253229], [-0.60834549, -0.63667341]]
X_aniso2 = np.dot(X2, transformation2)

fig = plt.figure(figsize=(8,8))
plt.scatter(X_aniso1[:, 0], X_aniso1[:, 1], c='r')
plt.scatter(X_aniso2[:, 0], X_aniso2[:, 1], c='g')

plt.show()
```

```
import numpy as np
X_aniso_total = np.append(X_aniso1,X_aniso2,axis=0)
fig = plt.figure(figsize=(8,8))
plt.scatter(X_aniso_total[:, 0], X_aniso_total[:, 1])

plt.show()
```
