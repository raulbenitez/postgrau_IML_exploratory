# Ejercicio: Comparar k-means y agrupamiento jerárquico 

En diferentes situaciones (número de clústers, distancia relativa, dispersión heterognénea, número de observaciones por cluster muy descompensado, forma alargada, etc)
 

```
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans, AgglomerativeClustering

from sklearn.datasets import make_blobs


n_samples = 1500
centers = [(0, 0), (5, 5),(10,10)]
random_state = 170


# Different variance
X_varied, y_varied = make_blobs(n_samples=n_samples,
                                cluster_std=[1.0, 2.5, 0.5],centers=centers,
                                random_state=random_state)

groupkmean = KMeans(n_clusters=3).fit_predict(X_varied)
groupaglomerative = AgglomerativeClustering(n_clusters=3).fit_predict(X_varied)
```

```
groupaglomerative.shape
```

```
groupkmean.shape
```

Representar cada situación:

```
plt.figure(figsize=(12, 12))



plt.subplot(121)
plt.scatter(X_varied[:, 0], X_varied[:, 1],c=groupkmean)
plt.title("kmeans")

plt.subplot(122)
plt.scatter(X_varied[:, 0], X_varied[:, 1],c=groupaglomerative)
plt.title("Clústers con diferente numero de observaciones")
plt.show()
```
