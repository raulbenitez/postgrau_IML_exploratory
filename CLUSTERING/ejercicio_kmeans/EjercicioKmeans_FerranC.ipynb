```
import seaborn as sns;
sns.set(style="ticks", color_codes=True)
mpg = sns.load_dataset("mpg")
mpg.head()
```

```
import pandas as pd
x1 = mpg[['mpg','cylinders','displacement','horsepower','weight','acceleration','model_year']]
y1 = mpg['origin']

x=pd.DataFrame(x1)
y=pd.DataFrame(y1)
```

```
g = sns.pairplot(mpg,hue="origin")
```

```
x_na = x.dropna()
print("Shape sin Nans {}, shape con Nans {}".format(x_na.shape,x.shape))
```

```
from sklearn.cluster import KMeans
kmeans = KMeans(n_clusters=3).fit_predict(x_na)

kmeans
```

```
import pandas as pd
mpg_na = mpg.dropna()
mpg_grouped = mpg_na
mpg_grouped['kmeans']= kmeans
g= sns.pairplot(mpg_grouped, hue='kmeans')
```

```
mpg_grouped.head(20)
```

```
kmeans_fit = KMeans(n_clusters=3).fit(x_na)
print('Etiquetas de asignacion a clusters {}'.format(kmeans_fit.labels_))
print("Posición de los centroides de los clusters {}".format(kmeans_fit.cluster_centers_))
print("Numero de iteraciones: {}".format(kmeans_fit.n_init))
```
