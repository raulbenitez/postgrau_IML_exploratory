```
import pandas as pd
import seaborn as sns
mpg = sns.load_dataset('mpg')
mpg=mpg.drop('cylinders',1)
mpg.head()
```

```
mpg_nonan = mpg.dropna()
mpg_num = mpg_nonan.select_dtypes(include='number')
mpg_num.head()
```

```
g = sns.pairplot(mpg_nonan, hue="origin")
```

```
X = mpg_num
y = mpg_nonan['origin'] 
```

```
from sklearn.cluster import KMeans
kmeans = KMeans(n_clusters=3).fit_predict(X)
```

```
mpg_grouped = mpg_num
mpg_grouped['kmeans']= kmeans
g= sns.pairplot(mpg_grouped, hue='kmeans')
```

```
gr = KMeans(n_clusters=3).fit(X)
print(gr.labels_)
print(gr.cluster_centers_)
print(gr.n_iter_)
```
