```
import pandas as pd 
import seaborn as sns 

mpg = sns.load_dataset("mpg")
mpg.drop(['cylinders', 'model_year'], axis='columns', inplace=True)
mpg_nonans= mpg.dropna ()
mpg_num= mpg_nonans.select_dtypes (include = 'number')


```

```
mpg_nonans.columns
```

```
X = mpg_num
y = mpg_nonans['origin'] 
```

```
g = sns.pairplot(mpg, hue="origin")
```

```
from sklearn.cluster import KMeans
group = KMeans(n_clusters=3).fit_predict(X)

X = mpg_num

mpg_grouped = mpg_nonans
mpg_grouped['kmeans'] = group
g = sns.pairplot(mpg_grouped, hue="kmeans")
```
