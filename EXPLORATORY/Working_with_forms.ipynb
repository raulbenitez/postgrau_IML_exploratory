Podemos pedir el path del archivo csv a cargar. Si no queremos que aparezca el código, tenemos que ir a propiedades de la Form y marcar 'Hide Code'

```
import pandas as pd
#@markdown ### Indica la ubicació del fitxer csv:
file_path = "/content/gdrive/My Drive/data_course/mpg_cars/auto-mpg.csv" #@param {type:"string"}
d = pd.read_csv(file_path)
```

```
import ipywidgets as widgets
colnms = list(d.columns)

var_picker = widgets.Dropdown(options=colnms, value='mpg')

var_picker
```

Ahora podemos utilizar la variable seleccionada para poder listar los valores de la columna correspondiente del dataframe de pandas:

```
d[var_picker.value].head()
```

O generar una gráfica a partir de la variable seleccionada

```
import matplotlib.pyplot as plt

fig = plt.figure(figsize=(10,6))
ax = plt.plot(d[var_picker.value])
plt.title('Evolució del mpg')
plt.show()
```
