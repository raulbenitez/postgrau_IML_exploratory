```
df2 = df_num_nonans.drop(columns=['is_canceled','is_repeated_guest','previous_cancellations','days_in_waiting_list', 'required_car_parking_spaces','adults','children', 'babies', 'previous_bookings_not_canceled', 'total_of_special_requests'])
df2.head()
```

```
import seaborn as sns
import matplotlib.pyplot as plt
sns.pairplot(df2)
plt.show()
```

```
X = df2

from sklearn.preprocessing import StandardScaler
XS = StandardScaler().fit_transform(X)
```

```
from sklearn import decomposition
n_variables = df2.shape[1]
pca = decomposition.PCA(n_components=n_variables).fit(XS)
print('varianza explicada acumulada ={}'.format(100*pca.explained_variance_ratio_.cumsum()))
```

```
Xproj = pca.transform(XS)
ndim_reducida = 7
XPCA = Xproj[:,0:ndim_reducida]
print('Dimensionalidad datos en espacio PCA reducido = {}'.format(XPCA.shape))
```

```
d = pd.DataFrame(XPCA,columns=['PCA1','PCA2','PCA3','PCA4','PCA5','PCA6','PCA7'])
sns.pairplot(d)
plt.show()
```

```
from sklearn.svm import SVC
from sklearn.datasets import load_digits
from sklearn.feature_selection import RFE
import matplotlib.pyplot as plt
import numpy as np
import matplotlib.pylab as plt
import seaborn as sns

df = pd.read_csv('hotel_bookings.csv')

df_nonan = df.dropna()
df_num_nonans = df_nonan.select_dtypes(include='number')

x= df_num_nonans
y = df_nonan['distribution_channel']

svc = SVC(kernel="linear", C=1)
rfe = RFE(estimator=svc, n_features_to_select=7, step=1)
rfe.fit(x, y)
print('Mantained features: {}'.format(rfe.support_))
print('Raking of the features: {}'.format(rfe.ranking_))
```
