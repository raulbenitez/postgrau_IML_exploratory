# Subplots in Python

## Using add_subplot

```python
import pandas as pd 
import seaborn as sns 
import matplotlib.pyplot as plt

mpg = sns.load_dataset("mpg")

fig = plt.figure(figsize=(10,10))
ax = fig.add_subplot(221)
ax = sns.lineplot(x="model_year", y="mpg", data=mpg)
ax.set_title('mpg vs model year')
ax1 = fig.add_subplot(222)
ax1 = sns.lineplot(x="model_year", y="weight", data=mpg)
ax1.set_title('weight vs model year')
ax2 = fig.add_subplot(223)
ax2 = sns.lineplot(x="cylinders", y="weight", data=mpg)
ax2.set_title('weight vs cylinders')
ax3 = fig.add_subplot(224)
ax3 = sns.lineplot(x="weight", y="acceleration",data=mpg)
ax3.set_title('acceleration vs weight')
plt.show()
```

## plt.subplots and shared axis

```python
# Create two subplots and unpack the output array immediately
f, (ax1, ax2) = plt.subplots(1, 2, sharey=True,figsize=(12,12))
sns.scatterplot(x="model_year", y="weight", data=mpg,ax=ax1)
ax1.set_title('Sharing Y axis')
sns.scatterplot(x="mpg", y="acceleration", data=mpg,ax=ax2)
ax2.set_title('Sharing Y axis')
plt.show()
```

## Matrix with axes 

```python
fig, axs = plt.subplots(2, 2,sharey=True,figsize=(12,12))
sns.scatterplot(y="mpg", x="weight", data=mpg,ax=axs[0, 0])
sns.scatterplot(y="mpg", x="acceleration", data=mpg,ax=axs[0, 1])
sns.scatterplot(y="mpg", x="horsepower", data=mpg,ax=axs[1, 0])
sns.scatterplot(y="mpg", x="model_year", data=mpg,ax=axs[1, 1])
plt.show()
```
