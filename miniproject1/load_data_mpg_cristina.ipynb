```
import pandas as pd
import seaborn as sns
mpg = sns.load_dataset("mpg")

mpg_num = mpg.select_dtypes(include='number')
mpg_num_nonans = mpg_num.dropna()

y = mpg['origin']
mpg_num_nonans.shape
```

```
g = sns.pairplot(mpg, hue="origin")
```

```
mpg_num_nonans.columns
```

```
x = mpg_num_nonans
y = mpg['origin']
```

```
from sklearn.preprocessing import StandardScaler
xs = StandardScaler().fit_transform(x)
```

```
from sklearn import decomposition
n_variables = mpg_num.shape[1]
pca = decomposition.PCA(n_components=n_variables).fit(xs)
```

```
print ('varianza acumulada = {}'.format (100*pca.explained_variance_ratio_.cumsum()))
```

```
Xproj = pca.transform(xs)
```

```
xPCA = Xproj[:,0:3]
```

```
print('Dimensionalidad PCA = {}'.format(xPCA.shape))
```
